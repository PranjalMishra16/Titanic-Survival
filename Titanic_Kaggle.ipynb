## **Extracting Features**

```python
#combining train and test dataset
train_test_data = [train_titanic, test_titanic]

#extracting titles from Name column.
for dataset in train_test_data:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.')
```

```python
train_titanic.head()
```

```python
pd.crosstab(train_titanic['Title'], train_titanic['Sex'])
```

We will now replace titles that aren't as common

```python
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col', \
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Other')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

train_titanic[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

```python
#Converting the titles into numeric form
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Other": 5}
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
```

```python
train_titanic.head()
```

```python
#converting sex into numeric form
for dataset in train_test_data:
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

```python
train_titanic.head()
```






---




```python
train_titanic.Embarked.value_counts()
```

```python
#replacing NaN values with S, since it has the largest number of passengers
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
```

```python
train_titanic.head()
```

```python
#Converting Embarked column into numeric form
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
```

```python
train_titanic.head()
```

```python
for dataset in train_test_data:
    age_avg = dataset['Age'].mean()#Finding mean of Age
    age_std = dataset['Age'].std()#Finding standard deviation
    age_null_count = dataset['Age'].isnull().sum() #Checking for null values ina ge

    age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
    #Replacing null values of age with random number
    #between mean of age - standard deviation of age to mean of age + standard deviation of age
    dataset.loc[np.isnan(dataset['Age']), 'Age'] = age_null_random_list #Changes values where missing
    dataset['Age'] = dataset['Age'].astype(int) #To convert data type

    train_titanic['AgeBand'] = pd.cut(train_titanic['Age'], 5)#Distributes the data in 5 equal sizes

print (train_titanic[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean())#Survival mean for each group
```

```python
train_titanic.head()
```

```python
#Changing age according to the age bands
for dataset in train_test_data:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4
```

```python
train_titanic.head()
```

```python
train_titanic['Fare'].isnull().sum()
```

```python
test_titanic['Fare'].isnull().sum()
```

```python
for dataset in train_test_data:
    dataset['Fare'] = dataset['Fare'].fillna(train_titanic['Fare'].median())
```

```python
train_titanic['FareBand'] = pd.qcut(train_titanic['Fare'], 4)#pd.qcut() is used for dividing a column of data
print (train_titanic[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean())
```

```python
train_titanic.head()
```

```python
#Changing age according to the age bands
for dataset in train_test_data:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
```

```python
train_titanic.head()
```

```python
#To see if families fare better in survival compared to a single traveller
for dataset in train_test_data:
    dataset['Family'] = dataset['SibSp'] +  dataset['Parch'] + 1

print (train_titanic[['Family', 'Survived']].groupby(['Family'], as_index=False).mean())
```

## **Feature Selection**

```python
#Dropping data we don't need for our model
features_drop = ['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Family']
train = train_titanic.drop(features_drop, axis=1)
test = test_titanic.drop(features_drop, axis=1)
train = train.drop(['PassengerId', 'AgeBand', 'FareBand'], axis=1)
```

```python
train.head()
```

```python
test.head()
```

```python
#Checking data
X_train = train.drop('Survived', axis=1)
y_train = train['Survived']
X_test = test.drop("PassengerId", axis=1).copy()

X_train.shape, y_train.shape, X_test.shape
```

# **Models**

```python
# Importing Classifier Modules
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

**Checking accuracies of models**


Logistic regression is a statistical method used for binary classification, which means it's used to predict the probability of a binary outcome (e.g., yes/no, true/false, 1/0). It's called "logistic" because it's based on the logistic function (also known as the sigmoid function), which maps any real-valued number to a value between 0 and 1. This makes it suitable for modeling
probabilities.

Prediction: After calculating the linear combination
z=b0+b1x1+b2x2+…+bnxn

the sigmoid function is applied to
z to obtain the predicted probability:
P(y=1|x)=σ(z)Where
P(y=1x) is the probability that the outcome is positive (class 1) given the input features x.

```python
#Checking accuracy of Logistic Regression Model
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_predicted= clf.predict(X_test)
acc_reg = round( clf.score(X_train, y_train) * 100, 2)
print ("Train Accuracy: " + str(acc_reg) + '%')
```

```python
#Checking accuracy of KNN Model

clf = KNeighborsClassifier(n_neighbors = 3)#3 closest data points are considered
clf.fit(X_train, y_train)#This trains the model using training data, model learns the relationship between features in both labels
y_pred_knn = clf.predict(X_test)#Predicts the target for test data (X_test)
acc_knn = round(clf.score(X_train, y_train) * 100, 2)#Calculates the accuracy of the model
print ("Train Accuracy: " + str(acc_knn) + '%')
```

K-Nearest Neighbors (KNN) is a simple machine learning algorithm used for both classification and regression tasks.
It works by finding the k most similar data points (neighbors) to a new data point and using the labels of those neighbors to predict the label of the new data point. The nearest neighbor is found using distance formulas,Euclidean Distance(d =√[(x2 – x1)2 + (y2 – y1)2]) Manhattan Distance(d = |x1 – y1| + |x2 – y2|), Minkowski Distance(( |X1 – Y1|p + |X2 – Y2|p + |X2 – Y2|p )1/p)
